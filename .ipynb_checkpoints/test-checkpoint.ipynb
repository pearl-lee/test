```python
def f(x):
    return np.log(x)
def f2(x):
    return np.exp(x)
x  = np.linspace(-5, 5, 100)

plt.plot(x, f(x), 'm', label='$logx$')
plt.plot(x, f2(x), 'b', label='$exp(x)$')
plt.plot(x, x, 'g', label='$x$')

plt.xlim(-3, 3); plt.ylim(-3, 3)
plt.xticks(np.arange(-3, 3))
plt.yticks(np.arange(-3, 3))
plt.legend()
plt.show()
```
